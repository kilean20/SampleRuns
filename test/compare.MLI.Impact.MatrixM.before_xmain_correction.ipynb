```python
import pImpactR as impact
np=impact.np
plt=impact.plot.plt
```

```python
beam = impact.getBeam()
beam.frequency = impact.data.clight/360
```

```python
ke = beam.kinetic_energy
mass = beam.mass
freq = beam.frequency
g = ke/mass + 1
bg = np.sqrt(g*g-1)
```

```python
def compare(lattice):
    impactM = impact.getTransferMap(beam,lattice)
    impactM.values[:5,-1]=-impactM.values[:5,-1]*mass*bg
    
    elemList,lattice = impact.impact2mli(beam,lattice)
    mapout = impact.MLI.getElem.ptm()
    elemList.append(mapout)
    Labor = impact.MLI.buildLabor(elemList[-1:])
    Labor.list.insert(0,'impactLattice')
    impact.MLI.writeInputfile(elemList,[lattice],Labor)
    impact.MLI.run()
    M,G = impact.MLI.readTransferMap()
    
    return impactM, M
```

# Drift

```python
elem = impact.getElem('drift')
elem.length = 0.2
elem.n_map = 1
elem.n_sckick = 20
M0,M1 = compare([elem])
```
