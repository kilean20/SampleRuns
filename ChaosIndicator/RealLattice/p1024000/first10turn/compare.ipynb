```python
import numpy as np
import pImpactR as impact
from copy import deepcopy as copy
import matplotlib.pyplot as plt
from math import ceil
import os
```

```python
beam,lattice = impact.readInputFile('test_IOTA.in')
ke = beam.kinetic_energy
mass = beam.mass
freq = beam.frequency
```

```python
chad   = impact.data.dictClass()
kilean = impact.data.dictClass()
```

# fort.24

```python
chad['X']   = impact.readRMS('x',fileLoc='./Chad/')
kilean['X'] = impact.readRMS('x',fileLoc='./Kilean/')
chad.n2 = np.argmin(np.abs(chad.X.s-360.0))
kilean.n2 = np.argmin(np.abs(chad.X.s-360.0))
chad.n1 = np.argmin(np.abs(chad.X.s-340.0))
kilean.n1 = np.argmin(np.abs(chad.X.s-340.0))
```

```python
plt.figure(figsize=(12,9))
plt.subplot(2,2,1)
plt.plot(chad.X.s[:chad.n2],chad.X.rms_x[:chad.n2],alpha=0.8)
plt.plot(kilean.X.s[:kilean.n2],kilean.X.rms_x[:kilean.n2],alpha=0.8)
plt.subplot(2,2,2)
plt.plot(chad.X.s[chad.n1:chad.n2],chad.X.rms_x[chad.n1:chad.n2],alpha=0.8)
plt.plot(kilean.X.s[kilean.n1:kilean.n2],kilean.X.rms_x[kilean.n1:kilean.n2],alpha=0.8,linestyle='--')
plt.subplot(2,2,3)
plt.plot(chad.X.s[:chad.n2],chad.X.rms_px[:chad.n2],alpha=0.8)
plt.plot(kilean.X.s[:kilean.n2],kilean.X.rms_px[:kilean.n2],alpha=0.8)
plt.subplot(2,2,4)
plt.plot(chad.X.s[chad.n1:chad.n2],chad.X.rms_px[chad.n1:chad.n2],alpha=0.8)
plt.plot(kilean.X.s[kilean.n1:kilean.n2],kilean.X.rms_px[kilean.n1:kilean.n2],alpha=0.8,linestyle='--')
```
