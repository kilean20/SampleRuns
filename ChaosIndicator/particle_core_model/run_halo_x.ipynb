```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
def run(fname):
    os.system('cp '+fname+' halo.in')
    os.system('./halo.x')
    return np.loadtxt('halo.out')
```

```python
data_a = run('halo_a.in')
plt.scatter(data_a[:,1],data_a[:,2],s=1)
```
