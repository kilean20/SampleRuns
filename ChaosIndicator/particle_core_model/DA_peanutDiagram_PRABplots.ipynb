# Particle-Core Model

Hamiltonians (normalized phase-space) for the particles and beam core are

$$
\begin{align}
H(x,p)&=\frac{p^{2}}{2}+\frac{x^{2}}{2}+\left(\eta^{2}-1\right)\times\begin{cases}
\frac{x^{2}}{2\mathcal{R}^{2}(t)} & |x|<\mathcal{R}(t)\\
\ln(x) & |x|\ge\mathcal{R}(t)
\end{cases}
\\
\mathcal{H}\left(\mathcal{R},\mathcal{P}\right)&=\frac{\mathcal{P}^{2}}{2}+\frac{\mathcal{R}^{2}}{2}+\frac{\eta^{2}}{2\mathcal{R}^{2}}+\left(\eta^{2}-1\right)\ln(\mathcal{R}) 
\end{align}
$$

where $\eta=\nu/\nu_{0}$ is the tune depression and the independent variable is in unit of $1/2\pi\nu_{0}$.
