```python
from IPython.display import Image,display
import matplotlib.pyplot as plt
from matplotlib.image import imread

def showImagesHorizontally(figsize,list_of_files,list_of_captions=None):
    fig = plt.figure(figsize=figsize)
    number_of_files = len(list_of_files)
    for i in range(number_of_files):
        a=fig.add_subplot(1,number_of_files,i+1)
        image = imread(list_of_files[i])
        plt.imshow(image)
        plt.axis('off')
        if list_of_captions != None:
            plt.title(list_of_captions[i])
    plt.tight_layout()
    
dTune = 4
path = './1024/'
extension = str(dTune)+'.png'
```

### Diff Histogram

FMA : tune difference
$$ \delta \equiv  \sqrt {\Delta\nu_x^2 + \Delta \nu_y^2 }$$

```python
Image(filename=path+'FMA.hist.Adiff.'+extension,width=700)
```

FB : phase-space difference

$$ X \equiv x - i(\beta_x p_x + \alpha_x x) $$

$$ \delta \equiv  \sqrt {\Delta |X|^2 + \Delta |Y|^2 }$$

```python
Image(filename=path+'FB.hist.Adiff.'+extension) 
```

### Dynamic Aperture

Log $\equiv log_e (\delta)$ 

Sig $\equiv \frac{1}{1+e^{-6 \frac{\delta-\bar{\delta}}{\sigma_\delta}}}$

```python
flist=[path+'FMA.DA.Log.A.'+extension,path+'FMA.DA.Sig.A.'+extension]
clist=['FMA.log','FMA.sigmoid']
showImagesHorizontally([10,4],flist)
flist=[path+'FB.DA.Log.A.'+extension,path+'FB.DA.Sig.A.'+extension]
clist=['FB.log','FB.sigmoid']
showImagesHorizontally([10,4],flist)
```

### Invariants

```python
flist=[path+'FMA.Inv.Log.A.'+extension,path+'FB.Inv.Log.A.'+extension,path+'FB.Inv.Sig.A.'+extension]
clist=['FMA.log','FB.log','FB.sigmoid']
showImagesHorizontally([16,8],flist,clist)
```

### Poincare

```python
flist=[path+'FMA.Poincare.X.'+extension,path+'FB.Poincare.X.'+extension]
clist=['FMA','FB']
showImagesHorizontally([10,5],flist,clist)
```
