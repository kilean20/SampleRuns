```python
from pImpactR import util
from pImpactR import opt
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
Mref = np.matrix([[1.0,0.0],[-1.0,1.0]])
```

```python
def Mtot(K):
    M = util.Mdrift(0.5)
    for i in range(len(K)):
        M=M*util.Mquad(K[i],0.1)*util.Mdrift(0.5)
    return M

def cost(K):
    M = Mtot(K)
    f = np.sum(np.multiply(M-Mref,M-Mref))
    for d in np.linspace(0.9,1.1,32):
        M = Mtot(K*d)
        f = f + np.sum(np.multiply(M-Mref,M-Mref))
    return f
```

```python
M1 = np.matrix([[1.0,.0],[-1.1,1.0]])
np.sum(np.multiply(M1-Mref,M1-Mref))
```

```python
with open('opt.result.pickle', 'rb') as f:
    result = pickle.load(f)
```

```python
bounds = [[-20.0,20.0]]*10
result = opt.differential_evolution(cost,bounds,ncore=32,popsize=2048,prev_result=result,
                                    mutation = [0.7,1.4], recombination = 0.6,
                                    disp=True,maxiter=2**20,maxtime=60*60*20)
```
