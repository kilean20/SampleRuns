```python
import numpy as np
import matplotlib.pyplot as plt
import pImpactR as impact
```

```python
EmitStr = ['7','14','28']
tunex = 0.59741947 # - 0.001
maxTurn = 100
_x = 0
_xSTR = 'X'
n_particles = 32000
```

```python
Me = impact.util.Me
ke = 100*1.0e6
energy = ke+Me
freq = 30*1.0e6
offset = 0.0003

g = energy/Me
bg = np.sqrt(g**2-1.0)

Emit  = [float(emit)*1.0e-9 for emit in EmitStr]
EmitN = [float(emit)*bg*1.0e-9 for emit in EmitStr]
```

```python
X=[]
N=[]
nMax=[]
i=0
for i in range(len(EmitStr)):
    X.append(np.load(_xSTR+'.ent.emit'+EmitStr[i]+'nm.npy'))
    N.append(np.arange(len(X[i]))*np.sqrt(EmitN[i]))
    nMax.append(int(maxTurn/np.sqrt(EmitN[i])))
    i=i+1

for i in range(len(Emit)):
    X[i]=X[i]*np.exp(-1j*np.angle(X[i][0]))
    
print(nMax)
print([len(X[i]) for i in range(len(Emit))])
print([X[i][0] for i in range(len(Emit))])
```

```python
fftX = []
Tune = []
for i in range(len(EmitStr)):
    fftX.append(np.fft.fft(X[i]))
    Tune.append(np.arange(0,1,1.0/len(X[i])))
```

# Raw Data

```python
plt.figure(figsize=(8,4))
plt.subplot(2,2,1)
for i in range(len(Emit)):
    plt.plot(N[i][:nMax[i]],np.abs(X[i][:nMax[i]])/np.abs(X[i][0]),alpha=0.7)
plt.xlabel(r'$n\sqrt{\epsilon_0}$',fontsize=12);
plt.ylabel(r'$|< z >| / |z_0|$',fontsize=12);
legend = [emit + ' nm' for emit in EmitStr]
plt.legend(legend)

plt.subplot(2,2,2)
for i in range(len(Emit)):
    plt.plot(N[i][:nMax[i]],np.abs(X[i][:nMax[i]])/np.abs(X[i][0]),alpha=0.7)
plt.xlabel(r'$n\sqrt{\epsilon_0}$',fontsize=12);
plt.ylabel(r'$|< z >| / |z_0|$',fontsize=12);
legend = [emit + ' nm' for emit in EmitStr]
plt.legend(legend)
plt.xlim(0,10)

plt.subplot(2,2,3)
for i in range(len(EmitStr)):
    plt.plot(Tune[i],np.real(fftX[i]),alpha=0.7)
plt.xlabel('tune',fontsize=12);
plt.ylabel(r'Re$\left[fft\left( z \right)\right]$',fontsize=12);
plt.legend(legend)

plt.subplot(2,2,4)
for i in range(len(EmitStr)):
    plt.plot(Tune[i],np.imag(fftX[i]),alpha=0.7)
plt.xlabel('tune',fontsize=12);
plt.ylabel(r'Im$\left[fft\left( z \right)\right]$',fontsize=12);
plt.legend(legend)

plt.tight_layout()
plt.savefig('Centroid.fft.ent.png',dpi=180,bbox_inches = "tight")
```

# dft zoom

```python
nSample = 1024
Tune= []
Amp = []
for i in range(len(Emit)):
    Tune.append(tunex + np.linspace(-5.0,5.0,nSample)*np.sqrt(float(Emit[i])))
    Amp.append(np.zeros(nSample,dtype=np.complex128))
    n=len(X[i])
    for k in range(nSample):
        Amp[i][k] = np.sum(X[i]*np.exp(-1j*2.0*np.pi*Tune[i][k]*np.arange(0,n)))
    Amp[i] = Amp[i]*np.sqrt(float(Emit[i]))*n_particles

pDataInit = impact.readParticleData(-111114,ke,Me,freq)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
for i in range(len(Emit)):
    plt.plot(np.linspace(-5.0,5.0,nSample),np.real(Amp[i])*1.3e1)
plt.hist((pDataInit[:,_x]-offset+0.00005)*4.6e3,bins=32,alpha=0.9,histtype ='step',color='k');
plt.hist((pDataInit[:,_x]-offset+0.00005)*4.6e3,bins=32,alpha=0.3,histtype ='bar');
#plt.xlim(-5.0,10.0)
plt.xlabel(r'$(\nu-\nu_0)/\sqrt{\epsilon_0}$',fontsize=12);
plt.legend(legend)
plt.grid()
plt.subplot(1,2,2)
for i in range(len(Emit)):
    plt.plot(np.linspace(-10.0,10.0,nSample),np.imag(Amp[i]))
#plt.xlim(-10.0,10.0)
plt.xlabel(r'$(\nu-\nu_0)/\sqrt{\epsilon_0}$',fontsize=12);
plt.legend(legend)
plt.grid()
plt.tight_layout()
plt.savefig('dft.ent.v0.png',dpi=180,bbox_inches = "tight")
```
